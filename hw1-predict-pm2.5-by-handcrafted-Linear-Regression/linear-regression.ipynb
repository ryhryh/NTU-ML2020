### train

```python
def process_one_day_one_feature(row):
    dts=pd.date_range(start=row['日期'].values[0], periods=24, freq='H', closed='left')
    feature=row['測項'].values[0]
    station=row['測站'].values[0]
    values=row.iloc[0,3:].values

    dummy=pd.DataFrame()
    dummy['dt']=dts
    dummy['station']=station
    dummy['feature']=feature
    dummy['value']=values
    dummy['value']=dummy['value'].apply(lambda x: str2float(x))
    return dummy
```

```python
# initialize
learning_rate=10
weight_num=x_train.shape[1]
W=np.zeros(weight_num).reshape(-1,1)
G_square_sum=np.zeros(weight_num).reshape(-1,1)
epochs=1000
eps=np.finfo(float).eps

# training
for i in range(epochs):
    # calculate gradients
    y_train_hat = x_train @ W
    y_train_delta = y_train_hat - y_train ###
    
    G = 2 * x_train.T @ y_train_delta
    G_square_sum += G ** 2
    sigma = np.sqrt(G_square_sum) + eps # prevent 0
    adaG = G / sigma
    
    # update weights
    W -= learning_rate * adaG

    # metrics
    y_valid_hat = x_valid @ W
    y_valid_delta = y_valid_hat - y_valid
    
    train_mse = np.mean(y_train_delta**2) 
    valid_mse = np.mean(y_valid_delta**2) 
    print(i, 'train_mse = %.3f'%(train_mse), 'valid_mse = %.3f'%(valid_mse))
```
